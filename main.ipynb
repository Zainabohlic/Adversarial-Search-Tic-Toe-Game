```python
import time

# Python3 program to find the next optimal move for a player
player, opponent = 'x', 'o'

# Global variable to count nodes
count = 0

# This function returns true if there are moves
# remaining on the board. It returns false if
# there are no moves left to play.
def isMovesLeft(board):
    for i in range(3):
        for j in range(3):
            if (board[i][j] == '_'):
                return True
    return False

def evaluate(b):
    # Checking for Rows for X or O victory.
    for row in range(3):
        if (b[row][0] == b[row][1] and b[row][1] == b[row][2]):
            if (b[row][0] == player):
                return 10
            elif (b[row][0] == opponent):
                return -10
    # Checking for Columns for X or O victory.
    for col in range(3):
        if (b[0][col] == b[1][col] and b[1][col] == b[2][col]):
            if (b[0][col] == player):
                return 10
            elif (b[0][col] == opponent):
                return -10
    # Checking for Diagonals for X or O victory.
    if (b[0][0] == b[1][1] and b[1][1] == b[2][2]):
        if (b[0][0] == player):
            return 10
        elif (b[0][0] == opponent):
            return -10
    if (b[0][2] == b[1][1] and b[1][1] == b[2][0]):
        if (b[0][2] == player):
            return 10
        elif (b[0][2] == opponent):
            return -10
    # Else if none of them have won then return 0
    return 0

# This is the minimax function. It considers all
# the possible ways the game can go and returns
# the value of the board
def minimax(board, depth, isMax):
    global count
    count += 1
    score = evaluate(board)
    # If Maximizer has won the game return his/her
    # evaluated score
    if (score == 10):
        return score
    # If Minimizer has won the game return his/her
    # evaluated score
    if (score == -10):
        return score
    # If there are no more moves and no winner then
    # it is a tie
    if (isMovesLeft(board) == False):
        return 0
    # If this maximizer's move
    if (isMax):
        best = -1000
        # Traverse all cells
        for i in range(3):
            for j in range(3):
                # Check if cell is empty
                if (board[i][j]=='_'):
                    # Make the move
                    board[i][j] = player
                    # Call minimax recursively and choose
                    # the maximum value
                    best = max(best, minimax(board, depth + 1, not isMax))
                    # Undo the move
                    board[i][j] = '_'
        return best
    # If this minimizer's move
    else:
        best = 1000
        # Traverse all cells
        for i in range(3):
            for j in range(3):
                # Check if cell is empty
                if (board[i][j] == '_'):
                    # Make the move
                    board[i][j] = opponent
                    # Call minimax recursively and choose
                    # the minimum value
                    best = min(best, minimax(board, depth + 1, not isMax))
                    # Undo the move
                    board[i][j] = '_'
        return best

# This will return the best possible move for the player
def findBestMove(board):
    global count
    start_time = time.time()
    count = 0
    bestVal = -1000
    bestMove = (-1, -1)
    # Traverse all cells, evaluate minimax function for
    # all empty cells. And return the cell with optimal
    # value.
    for i in range(3):
        for j in range(3):
            # Check if cell is empty
            if (board[i][j] == '_'):
                # Make the move
                board[i][j] = player
                # compute evaluation function for this
                # move.
                moveVal = minimax(board, 0, False)
                # Undo the move
                board[i][j] = '_'
                # If the value of the current move is
                # more than the best value, then update
                # best/
                if (moveVal > bestVal):
                    bestMove = (i, j)
                    bestVal = moveVal
    print("The value of the best Move is:", bestVal)
    print("Number of nodes visited:", count)
    print("Execution time:", time.time() - start_time, "seconds")
    print()
    return bestMove

# Driver code

# board = [
# [ 'x', 'o', 'x' ],
# [ 'o', 'o', 'x' ],
# [ '_', '_', '_' ]
# ]

# board = [    ['x', 'o', '_'],
#     ['o', 'o', 'x'],
#     ['_', 'x', '_']
# ]
board = [['x', '_', '_'],
['o', '_', '_'],
['_', 'o', 'x']]

# board = [    ['o', '_', 'x'],
#     ['_', 'x', '_'],
#     ['o', 'x', '_']
# ]

bestMove = findBestMove(board)
print("The Optimal Move is:")
print("ROW:", bestMove[0], "COL:", bestMove[1])
```

```python
import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_win(board, player):

    for i in range(3):
        if all(board[i][j] == player for j in range(3)) or all(board[j][i] == player for j in range(3)):
            return True

    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

def get_available_moves(board):
    moves = []
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                moves.append((i, j))
    return moves

def evaluate_state(board, player):
    opponent = 'o' if player == 'x' else 'x'
    player_wins = 0
    opponent_wins = 0

    for i in range(3):
        player_row_count = board[i].count(player)
        opponent_row_count = board[i].count(opponent)
        player_col_count = [board[j][i] for j in range(3)].count(player)
        opponent_col_count = [board[j][i] for j in range(3)].count(opponent)
        if player_row_count == 2 or player_col_count == 2:
            player_wins += 1
        elif opponent_row_count == 2 or opponent_col_count == 2:
            opponent_wins += 1

    player_diag1_count = [board[i][i] for i in range(3)].count(player)
    opponent_diag1_count = [board[i][i] for i in range(3)].count(opponent)
    player_diag2_count = [board[i][2 - i] for i in range(3)].count(player)
    opponent_diag2_count = [board[i][2 - i] for i in range(3)].count(opponent)
    if player_diag1_count == 2 or player_diag2_count == 2:
        player_wins += 1
    elif opponent_diag1_count == 2 or opponent_diag2_count == 2:
        opponent_wins += 1

    return player_wins - opponent_wins

def weak_ai_move(board, player, excluded_tile):
    available_moves = get_available_moves(board)
    available_moves = [move for move in available_moves if move != excluded_tile]
    return random.choice(available_moves)

def play_game():
    board = [['_', '_', '_'],
             ['_', '_', '_'],
             ['_', '_', '_']]
    players = ['x', 'o']
    random_tile = random.choice([(i, j) for i in range(3) for j in range(3)])
    print("Weak player won't use tile:", random_tile)
    weak_player = players[random.randint(0, 1)]
    print("Weak player:", weak_player)
    strong_player = 'x' if weak_player == 'o' else 'o'
    print("Strong player:", strong_player)

    turn = 0
    while True:
        print_board(board)
        current_player = players[turn % 2]
        if current_player == weak_player:
            move = weak_ai_move(board, current_player, random_tile)
        else:
            while True:
                move = tuple(map(int, input(f"Enter move for {current_player} (row col): ").split()))
                if board[move[0]][move[1]] == '_':
                    break
                else:
                    print("Invalid move! Try again.")

        board[move[0]][move[1]] = current_player

        if check_win(board, current_player):
            print_board(board)
            print(f"Player {current_player} wins!")
            break
        elif len(get_available_moves(board)) == 0:
            print_board(board)
            print("It's a tie!")
            break

        turn += 1

# Play the game
play_game()
```
